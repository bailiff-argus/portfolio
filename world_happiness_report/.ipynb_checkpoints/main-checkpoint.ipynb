```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

plt.rcParams['figure.dpi'] = 300
sns.set_theme()
sns.set_style('white')
```

```python
world_happiness_df = pd.read_csv("./data/2016.csv")

# for the purpose of making this case simpler, let's drop the confidence intervals
world_happiness_df.drop(["Lower Confidence Interval", "Upper Confidence Interval"], axis=1, inplace=True)

# the "dystopia residual" is a column of unexplained components and cannot be used for correlation analysis
world_happiness_df.drop("Dystopia Residual", inplace = True, axis = 1)

# happiness rank is basically another index, so let's set it as one
world_happiness_df.set_index("Happiness Rank", inplace=True)


world_happiness_df.head(10)
```

## Initial Look

We can immediately create a correlation matrix and examine the most correlated parameters.

```python
corr_matrix = world_happiness_df.corr()

fig, ax = plt.subplots(figsize = (11, 9))
cmap = sns.diverging_palette(230, 110, as_cmap=True)
sns.heatmap(data = corr_matrix, 
            cmap = cmap, vmax = 1, center = 0.5, linewidth = 0.3, square = True
           )
```

## Exploring different variables

It may be safe to assume that different variables are going to be affecting the happiness score to different degrees. Let's see how much they correlate with the happiness score.

### Region

```python
world_happiness_df.loc[:, ['Region', 'Happiness Score']].groupby('Region').agg(['mean', 'std']).sort_values(
    by = ("Happiness Score", "mean"), ascending = False
)
```

We can see that in different regions happiness scores can be very different. Notice that generally the regions that are known as "richer" have higher happiness scores.

### GDP

Notice that the highest standard deviations are observed in regions where both "rich" and "poor" countries are located, most prominent example being Middle East and Northern Africa. In that region there are both rich countries with oil-based economies and poor ones reliant on e.g. tourism as a main source of income.

Let's see how much of a role GDP plays here, and how correlated the standard deviations are.

```python
world_happiness_df.loc[:, ['Country', 'Happiness Score', 'Economy (GDP per Capita)']].sort_values(
    by = ("Happiness Score", "mean"), ascending = False
)
```

At first glance, happiness scores of regions seem to be correlated with the GDPs rather strongly, while the standard deviations don't. Let's examine the relationship visually.

```python
gdp_by_regions_df = world_happiness_df.loc[:, ['Region', 'Happiness Score', 'Economy (GDP per Capita)']].groupby(
    'Region'
).agg(['mean', 'std'])

sns.scatterplot(data = gdp_by_regions_df,
               x = ("Happiness Score", "mean"), y = ("Economy (GDP per Capita)", "mean"))
```

```python
sns.scatterplot(data = gdp_by_regions_df,
               x = ("Happiness Score", "std"), y = ("Economy (GDP per Capita)", "std"))
```

Means seem to be strongly positively correlated, while standard deviations show a weak positive linear relationship.

```python
happiness_score_df = gdp_by_regions_df.loc[:, ('Happiness Score', 'mean')]
gdp_per_capita_df = gdp_by_regions_df.loc[:, ('Economy (GDP per Capita)', 'mean')]

happiness_score_df.corr(gdp_per_capita_df)
```
