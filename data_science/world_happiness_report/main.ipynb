# The World Happiness 

We're going to find the happiest region on the planet based on freely available data.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

plt.rcParams['figure.dpi'] = 300
sns.set_theme()
sns.set_style('white')
```

```python
world_happiness_df = pd.read_csv("./data/2016.csv")

world_happiness_df.head()
```

```python
#let's change column names to something that's easier to work with
world_happiness_df.rename(
    columns = {
    'Country': 'country',
    'Region': 'region',
    'Happiness Rank': 'hap_rank',
    'Happiness Score': 'hap_score',
    'Economy (GDP per Capita)': 'gdp_per_cap',
    'Family': 'family',
    'Health (Life Expectancy)': 'health',
    'Freedom': 'freedom',
    'Trust (Government Corruption)': 'gov_trust',
    'Generosity': 'generosity',
}, inplace = True)


# happiness rank is basically another index, so let's set it as one
world_happiness_df.set_index("hap_rank", inplace=True)

# for the purpose of making this case simpler, let's drop the confidence intervals
world_happiness_df.drop(["Lower Confidence Interval", "Upper Confidence Interval"], axis=1, inplace=True)

# the "dystopia residual" is a column of unexplained components and cannot be used for correlation analysis
world_happiness_df.drop("Dystopia Residual", inplace = True, axis = 1)


world_happiness_df.head(10)
```

```python
world_happiness_df.info()
```

```python
world_happiness_df.describe()
```

## Initial Look

We can immediately create a correlation matrix and examine the most correlated parameters.

We have to note, however, that a negative correlation is very unlikely because of how the parameters work. Every single one of them shows how many points out of total happiness score are attributed to a certain variable.

However, we can still find some relationships. The question we're going to be asking ourselves is the following.

Does higher contribution of a given variable generally mean proportionate happiness growth?

```python
corr_matrix = world_happiness_df.corr()

corr_matrix
```

```python
fig, ax = plt.subplots(figsize = (12, 9))
cmap = sns.diverging_palette(230, 110, as_cmap=True)
g = sns.heatmap(data = corr_matrix, 
            cmap = cmap, vmax = 1, center = 0.5, linewidth = 0.5, square = True
    )
```

It's visible that high contribution of some parameters (like GDP per capita, or health) are strongly correlated with happiness scores, while others can be correlated comparatively weakly.

Let's make a "strong correlation group": GDP per capita, family, and health;<br>
and a "weak correlation group": freedom, government trust, and generosity.

Additionally, GDP per capita and health seem to correlate strongly as well.

## Exploring non-numerical variables

The region variable could not be added to correlation matrix because of its non-numerical nature. Let's see if the relationships established above hold for regions in general.

### Region

```python
world_happiness_df.loc[:, ['region', 'hap_score']].groupby('region').agg(['mean', 'std']).sort_values(
    by = ("hap_score", "mean"), ascending = False
)
```

We can see that in different regions happiness scores can be very different. Notice that generally the regions that are known as "richer" have higher happiness scores.

### Score study by region

Notice that the highest standard deviations are observed in regions where both "rich" and "poor" countries are located, most prominent example being Middle East and Northern Africa. In that region there are both rich countries with oil-based economies and poor ones reliant on e.g. tourism as the main source of income.

Let's examine the different parameters when the countries are grouped by their respective regions.

```python
region_happiness_df = world_happiness_df.groupby('region').agg(['mean'])

region_happiness_df.sort_values(('hap_score', 'mean'), ascending=False)
```

Let's see if region-based grouping gives us any valuable insights.

```python
reg_corr_matrix = region_happiness_df.corr()

reg_corr_matrix
```

```python
fig, ax = plt.subplots(figsize = (12, 9))

cmap = sns.diverging_palette(220, 110, as_cmap=True)
sns.heatmap(data = reg_corr_matrix, 
            cmap = cmap, vmax = 1, center = 0.5, linewidth = 0.3, square = True,
            ax = ax
    )

ax.set_xlabel(None)
ax.set_ylabel(None)
```

According to this heatmap, the happier regions generally have higher impacts from each of the variables. Additionally, the "weaker" correlations from before seem to be stronger here.

GDP per capita correlates very strongly with health scores. This might suggest that high GDP per capita and family scores are the most indicative of an overall happy country.

Now let's reach the initial goal and rank the regions.
